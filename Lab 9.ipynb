```python
#3 Get all films with ARMAGEDDON in the title.
```

```python
question3 = "select title from sakila.film where title like '%%ARMAGEDDON%%'"
data3 = pd.read_sql_query(question3, engine)
data3

```

```python
#4 Get 10 the longest films.
```

```python
question4 = 'SELECT * FROM sakila.film ORDER BY length DESC LIMIT 10'
data4 = pd.read_sql_query(question4, engine)
```

```python
#5. How many films include Behind the Scenes content?
```

```python
question5 = "select * FROM sakila.film where special_features like '%%Behind the Scenes%%'"
data5 = pd.read_sql_query(question5, engine)
```

```python
data5
```

```python
#6. Which kind of movies (rating) have a mean duration of more than two hours?
```

```python
question6 = "SELECT rating,round(avg(length),2) from sakila.film group by rating having round(avg(length),2)>120"
data6 = pd.read_sql_query(question6, engine)
data6
```

```python
#7. Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.
```
